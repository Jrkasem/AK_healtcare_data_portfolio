```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the CSV file
data = pd.read_csv('C:\\Users\\Dell\\Downloads\\combined data hh.csv')


#1 Create a bar plot for the distribution of practice types

# Extract the responses for the "Type of Practice (check all that apply): - Selected Choice" question
practice_types = data['Q1']

# Count the occurrences of each practice type
practice_counts = practice_types.str.split(',').explode().str.strip().value_counts()

# Create a bar plot for the distribution of practice types
plt.figure(figsize=(10, 6))
sns.barplot(x=practice_counts.index, y=practice_counts.values)
plt.xlabel('Practice Type')
plt.ylabel('Number of Responses')
plt.title('Distribution of Practice Types')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()



screening_responses = data['Q2']

#2 Count the occurrences of each response
screening_counts = screening_responses.value_counts()

# Create a pie chart for the distribution of screening responses
plt.figure(figsize=(8, 8))
plt.pie(screening_counts, labels=screening_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Universal Screening for TB Testing')
plt.show()


#3 Extract the responses for the "Which TB test do you use/order?" question
tb_test_responses = data['Q3']

# Count the occurrences of each TB test
tb_test_counts = tb_test_responses.value_counts()

# Create a bar plot for the distribution of TB tests used or ordered
plt.figure(figsize=(10, 6))
sns.barplot(x=tb_test_counts.index, y=tb_test_counts.values)
plt.xlabel('TB Test')
plt.ylabel('Number of Responses')
plt.title('Distribution of TB Tests Used/Ordered')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

#4 Extract the responses for the "Which of the following risk factors for TB infection and/or risk for progression to TB disease prompts you to order a TB test for a patient? Check all that apply. - Selected Choice" question
risk_factors_responses = data["Q4"]

# Combine all responses into a single list
all_responses = risk_factors_responses.str.cat(sep=',').split(',')

# Count the occurrences of each risk factor
risk_factors_counts = pd.Series(all_responses).value_counts()

# Create a horizontal bar plot for the distribution of risk factors that prompt respondents to order a TB test for a patient
plt.figure(figsize=(10, 6))
sns.barplot(x=risk_factors_counts.values, y=risk_factors_counts.index, palette='viridis')
plt.xlabel('Number of Responses')
plt.ylabel('Risk Factors')
plt.title('Risk Factors Prompting TB Test Order')
plt.tight_layout()
plt.show()


#5 # Extract the responses for the "On average, how many TB tests do you order or perform annually?" question
tb_tests_responses = data["Q5"]

# Create a count plot
plt.figure(figsize=(10, 6))
sns.countplot(x=tb_tests_responses, palette='viridis')
plt.xlabel('Number of TB Tests')
plt.ylabel('Count')
plt.title('Distribution of TB Tests Ordered or Performed Annually')
plt.tight_layout()

plt.show()











#6 Extract the responses for the "If a patient presents with any of the risk factors in Question 15 and has a history of receiving the Bacillus Calmette-Guérin (BCG) vaccine, do you?" question
bcg_vaccine_responses = data["Q6"]

# Count the occurrences of each action taken
bcg_vaccine_counts = bcg_vaccine_responses.value_counts()

# Create a bar plot for the distribution of actions taken when patients have risk factors and a history of BCG vaccine
plt.figure(figsize=(10, 6))
sns.barplot(x=bcg_vaccine_counts.index, y=bcg_vaccine_counts.values)
plt.xlabel('Actions Taken')
plt.ylabel('Number of Responses')
plt.title('Actions Taken for Patients with Risk Factors and BCG Vaccine History')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

#7 Extract the responses for the "If a patient has a new positive TB test (TST or IGRA), what action(s) do you take? Check all that apply. - Selected Choice" question
positive_tb_test_responses = data["Q7"]

# Combine all responses into a single list
all_responses = positive_tb_test_responses.str.cat(sep=',').split(',')

# Count the occurrences of each action
action_counts = pd.Series(all_responses).value_counts()

# Create a horizontal bar plot for the distribution of actions taken when a patient has a new positive TB test
plt.figure(figsize=(10, 6))
sns.barplot(x=action_counts.values, y=action_counts.index, palette='viridis')
plt.xlabel('Number of Responses')
plt.ylabel('Actions Taken')
plt.title('Actions Taken for Patients with a New Positive TB Test')
plt.tight_layout()
plt.show()

#8
ltbi_actions_responses = data["Q8"]

# Combine all responses into a single list
all_responses = ltbi_actions_responses.str.cat(sep=',').split(',')

# Count the occurrences of each action
action_counts = pd.Series(all_responses).value_counts()

# Create a horizontal bar plot for the distribution of actions taken in LTBI scenario
plt.figure(figsize=(10, 6))
sns.barplot(x=action_counts.values, y=action_counts.index, palette='viridis')
plt.xlabel('Number of Responses')
plt.ylabel('Actions Taken')
plt.title('Actions Taken for Patients with Positive TB Test and Negative Chest X-ray (LTBI)')
plt.tight_layout()
plt.show()

#9
ltbi_info_responses = data["Q9"]

# Combine all responses into a single list
all_responses = ltbi_info_responses.str.cat(sep=',').split(',')

# Count the occurrences of each option
option_counts = pd.Series(all_responses).value_counts()

# Create a bar plot for the distribution of preferred ways to receive information about LTBI
plt.figure(figsize=(10, 6))
sns.barplot(x=option_counts.values, y=option_counts.index, palette='viridis')
plt.xlabel('Number of Responses')
plt.ylabel('Preferred Ways to Receive Information')
plt.title('Preferred Ways to Receive Information about LTBI Screening, Testing, and Treatment')
plt.tight_layout()
plt.show()

#Q10
# Extract the responses for the "Would you, or a representative from your practice, be interested in sharing additional feedback as we develop LTBI resources? If yes, my preferred method of contact is" question
contact_method_responses = data["Q10"]

# Count the occurrences of each contact method
contact_method_counts = contact_method_responses.value_counts()

# Create a pie chart for the distribution of preferred methods of contact
plt.figure(figsize=(8, 8))
plt.pie(contact_method_counts, labels=contact_method_counts.index, autopct='%1.1f%%', colors=sns.color_palette('viridis'))
plt.title('Preferred Methods of Contact for Sharing Feedback on LTBI Resources')
plt.axis('equal')
plt.show()
```
